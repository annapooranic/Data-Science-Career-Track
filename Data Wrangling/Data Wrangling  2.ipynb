```python
youtube_data['author3'] = youtube_data['author3'].fillna(youtube_data['author2'])
youtube_data['author3'] = youtube_data['author3'].fillna(youtube_data['author1'])
youtube_data['talk3']=youtube_data['talk3'].str.replace('withenglishsubtitles','')
youtube_data['talk3']=youtube_data['talk3'].str.replace(',','').str.replace('?','').str.replace('|','').str.replace(':','').str.replace('"','').str.replace('!','').str.replace('-','')
```

```python
youtube_data['author3'] = youtube_data['author3'].str.strip()
youtube_data['author3'] = youtube_data['author3'].str.lower()
#ted_data['main_speaker'] =

ted_data['main_speaker'] = ted_data['main_speaker'].str.lower()
ted_data['main_speaker'] = ted_data['main_speaker'].str.strip()
```

```python
import nltk
from nltk import word_tokenize 
youtube_data['tokens_u']= youtube_data['talk3'].apply(nltk.word_tokenize)
ted_data['tokens']=ted_data['ted_title'].apply(nltk.word_tokenize)
youtube_data['tokens']=youtube_data['tokens_u'].str.slice(start=2,stop=None,step=None)
```

```python
youtube_data['sp_tokens']= youtube_data['author3'].apply(nltk.word_tokenize)
ted_data['sp_tokens']=ted_data['main_speaker'].apply(nltk.word_tokenize)
```

```python
from collections import Counter
youtube_data['y_talk_co'] = youtube_data.apply(lambda g:Counter(g['tokens_u']),axis=1)

ted_data['t_talk_co'] = ted_data.apply(lambda g:Counter(g['tokens']),axis=1)

youtube_data['y_sp_co'] = youtube_data.apply(lambda g:Counter(g['sp_tokens']),axis=1)
ted_data['t_sp_co'] = ted_data.apply(lambda g:Counter(g['sp_tokens']),axis=1)
```

```python
merge_outer = pd.merge(ted_data,youtube_data,left_on='name_slice',right_on='talk_slice',how='outer',indicator=True)
ted_only = merge_outer[merge_outer['_merge'] == 'left_only']
youtube_only = merge_outer[merge_outer['_merge'] == 'right_only']
```

```python
from scipy.spatial.distance import cosine
def second_function(ted_talk,youtube_talk,ted_speaker,youtube_speaker):
   # print(ted_talk,youtube_talk,ted_speaker,youtube_speaker)
    speak_concat = list(ted_speaker.keys()| youtube_speaker.keys())
    c_vect = [ted_speaker.get(word, 0) for word in speak_concat]       
    d_vect = [youtube_speaker.get(word, 0) for word in speak_concat]
   # print(c_vect,d_vect)
    cos1 = 1 - cosine(c_vect,d_vect)
   # print(cos1)
    if cos1 > 0.6:
       # print('enter')
        words  = list(ted_talk.keys() | youtube_talk.keys())
        e_vect = [ted_talk.get(word, 0) for word in words]
        f_vect = [youtube_talk.get(word, 0) for word in words]
        cost = 1 - cosine(e_vect,f_vect)
       # print(cost)
        return cost
    
        
        
```

```python
result = ted_only.apply(lambda x:youtube_only.apply(lambda y: second_function(x['t_talk_co'],y['y_talk_co'],x['t_sp_co'],y['y_sp_co']),axis =1),axis=1).rename(columns = youtube_only['youtube_title'],
                                                                                          index = ted_only['name'])
```
